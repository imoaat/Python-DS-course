```python
# Сделал 35 вариант, т.к. 14 варианта нет
# Дано целое число. Если оно является положительным, то прибавить к нему 1
# Если отрицательным, то вычесть из него 2; если нулевым, то заменить его на 10
while True:  # обработка исключений
    try:
        a = int(input('Введите целое число: '))
        break
    except ValueError:
        print('Введено неверно')

if a > 0:  # проверка условия a>0
    a += 1
elif a < 0: # проверка условия a<0
    a -= 2
else: # проверка условия a=0
    a = 10
print(a)
```

```python
#  Дано вещественное число A и целое число N (>0).
#  Используя один цикл, вывести все целые степени числа A от 1 до N.
i = 1
while True:  # обработка исключений
    try:
        a = float(input('Введите вещественное число: '))
        n = int(input('Введите целое число больше нуля: '))
        if n < 0:  # проверка условия n>0
            print('Второе число меньше нуля!')
        else:
            break

    except ValueError:
        print('Введено неверно')

while i <= n:  # цикл
    s = a ** i
    i += 1  # счётчик
    s = float(s)
    print('Ваше число в степени {} :'.format(i-1),s)
```

```python

































# С помощью функций получить вертикальную и горизонтальную линии.
# Линия проводится многократной печатью символа.
# Заключить слово в рамку из полученных линий.

# Создание из символа горизонтальной линии
def horizontalLine(count):
    return '-' * (count + 4)


# Создание двух вертикальных символов и последующие заключение в них текста
def verticalLine(text):
    return ('| {} |').format(text)


# Ввод значений и создание переменных
text = str(input('Введите текст: '))
textLen = len(text)
horLine = horizontalLine(textLen)
verLine = verticalLine(text)

# Вывод результатов
print(horLine)
print(verLine)
print(horLine)
```
